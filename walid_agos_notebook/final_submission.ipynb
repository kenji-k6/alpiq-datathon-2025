```python
import pandas as pd

# Read the Excel file
df = pd.read_excel("datasets2025/my_dataframe.xlsx")

# Columns to remove
columns_to_drop = [
    'is_holiday', 'spv', 'temp', 'temperature_2m', 'relative_humidity_2m',
    'cloudcover', 'precipitation', 'hour', 'day', 'month', 'year',
    'day_of_week', 'quarter', 'is_weekend', 'hour_of_day_category'
]

# Find and add any columns with these patterns
for col in df.columns:
    if 'VALUEMWHMETERINGDATA_customerES_' in col or 'INITIALROLLOUTVALUE_customerES_' in col:
        columns_to_drop.append(col)

# Drop the columns (only if they exist in the dataframe)
columns_to_drop = [col for col in columns_to_drop if col in df.columns]
df = df.drop(columns=columns_to_drop)

# Save the cleaned dataframe
df
```

```python
# Create a dictionary to map old column names to new column names
rename_dict = {}

# Loop through each column
for col in df.columns:
    # Check if the column follows the pattern "customerES_X"
    if col.startswith("customerES_"):
        # Extract the customer number
        customer_num = col.split("_")[1]
        # Create new column name with the desired prefix
        new_col_name = f"VALUEMWHMETERINGDATA_customerES_{customer_num}"
        # Add to rename dictionary
        rename_dict[col] = new_col_name

# Rename the columns
df = df.rename(columns=rename_dict)

# Save the updated dataframe
df.to_excel("my_dataframe_renamed.xlsx", index=False)

print(f"Renamed {len(rename_dict)} columns from 'customerES_X' to 'VALUEMWHMETERINGDATA_customerES_X'")
```
